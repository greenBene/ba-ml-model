```python
import pandas as pd
import numpy as np
import datetime as dt
import os
import time
```

```python
data_dir = 'geolife-data/Prepared'
dirlist = os.listdir(data_dir)

data = pd.DataFrame()

for dir in dirlist[1:2]:
    print(dir)
    data_temp = pd.read_csv(data_dir + '/' + dir).drop(columns=["Unnamed: 0"])
    data = pd.concat([data, data_temp])
```

```python
%matplotlib inline
```

```python
df = data[data["trans_trip"] == 5]
import folium   # (https://pypi.python.org/pypi/folium)
mymap = folium.Map(width=400,height=300, location=[ df.latitude.mean(), df.longitude.mean() ], zoom_start=14)
#folium.PolyLine(df[['latitude','longitude']].values, color="red", weight=2.5, opacity=1).add_to(mymap)

for coord in df[['latitude','longitude']].values:
    folium.CircleMarker(location=[coord[0],coord[1]], radius=1,color='blue').add_to(mymap)
```
